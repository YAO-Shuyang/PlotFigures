```python
import numpy as np
import umap
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import networkx as ntx
from mylib.statistic_test import *

code_id = '0849 - Visualize UMAP and Persistent Homology'
loc = os.path.join(figpath, 'Dsp', code_id)
mkdir(loc)

# Example dataset
X, _ = make_blobs(n_samples=20, centers=2, random_state=42, n_features=1)

# UMAP nearest neighbor graph
umap_obj = umap.UMAP(n_neighbors=10, min_dist=0.3, n_components=2, random_state=42)
umap_obj.fit(X)

# Get the graph
graph = umap_obj.graph_

# Visualize the graph using NetworkX
G = ntx.Graph(graph)
# Precompute the positions using a layout
pos = ntx.spring_layout(G, seed=42)  # Use a fixed seed for reproducibility

def plot(size):
    plt.figure(figsize=(2, 4))
    ax = Clear_Axes(plt.axes())
    # Draw edges (fully opaque)
    # Draw nodes (transparent)
    ntx.draw_networkx_nodes(
        G,
        pos=pos,  # Use the same layout as edges
        node_size=size,  # Enlarge nodes
        node_color="#D1C3E0",  # Node color
        alpha=0.8,  # Set transparency for nodes
        edgecolors='black',
        linewidths=0.5,
    )
    ntx.draw_networkx_edges(
        G,
        pos=pos,  # Position nodes
        edge_color="black",  # Edge color
        alpha=1.0,  # Fully opaque edges,
        width=0.5
    )
    plt.savefig(join(loc, f"Persistent Homology Size {size} .png"), dpi=600)
    plt.savefig(join(loc, f"Persistent Homology Size {size} .svg"), dpi=600)
    plt.tight_layout()
    plt.show()
    
plot(100)
plot(250)
plot(500)
plot(1000)
plot(2000)
```
