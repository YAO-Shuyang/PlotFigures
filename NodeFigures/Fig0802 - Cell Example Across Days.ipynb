```python
def plot_cell_examples_across_days(index_map, traces, save_loc):
    assert len(traces) == index_map.shape[0]
    
    adds = np.linspace(0, 600, 7)
    last_time = np.array([trace['correct_time'][-1] for trace in traces])
    last_time = np.concatenate([[0], last_time[:-1]])
    tot_additions = adds + np.cumsum(last_time)
    
    colors = [
        DSPPalette[0], DSPPalette[1], DSPPalette[2], DSPPalette[3], DSPPalette[0],
        DSPPalette[0], DSPPalette[4], DSPPalette[5], DSPPalette[6], DSPPalette[0]
    ]
    
    fig = plt.figure(figsize=(4, 4*7))
    ax = Clear_Axes(plt.axes(), close_spines=['top', 'right'], ifxticks=True, ifyticks=True)
    for cell in np.arange(index_map.shape[1]):
        print(f"Plotting Cell {index_map[:, cell]} at column {cell}...")
        y_ticks = []
        y_labels = []
        for d, trace in enumerate(traces):
            if index_map[d, cell] == 0:
                continue
        
            for i in range(10):
                ax = LocTimeCurveAxes(
                    ax,
                    behav_time=trace[f'node {i}']['ms_time_behav'] + tot_additions[d],
                    behav_nodes=spike_nodes_transform(trace[f'node {i}']['spike_nodes'], 12),
                    spikes=trace[f'node {i}']['Spikes'][index_map[d, cell]-1, :],
                    spike_time=trace[f'node {i}']['ms_time_behav'] + tot_additions[d],
                    maze_type=1,
                    line_kwargs={'linewidth': 0.4, 'color': colors[i]},
                    bar_kwargs={'markeredgewidth': 0.5, 'markersize': 2, 'color': 'k'},
                    is_include_incorrect_paths=True
                )[0]
            
            t1 = trace['node 4']['ms_time_behav'][-1]/1000 + tot_additions[d]/1000
            t2 = trace['node 5']['ms_time_behav'][0]/1000 + tot_additions[d]/1000
            t3 = trace['node 9']['ms_time_behav'][-1]/1000 + tot_additions[d]/1000
            y_ticks += [0, t1, t2, t3]
            y_labels += [0, t1, 0, t3-t2]
        
        ax.set_yticks(y_ticks, y_labels)
        plt.savefig(os.path.join(save_loc, f'cell_{cell+1}.png'), dpi=600)
        plt.savefig(os.path.join(save_loc, f'cell_{cell+1}.svg'), dpi=600)
        ax.clear()
            
save_loc = os.path.join(loc, f"{mouse}")
mkdir(save_loc)
plot_cell_examples_across_days(index_map, traces, save_loc)
        
```
