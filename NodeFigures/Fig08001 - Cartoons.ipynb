```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 10, 1000)

def gaussian(x0=0, sigma=2, pi=np.pi):
    return 1 / (sigma * np.sqrt(pi * 2)) * np.exp(- (x-x0)**2 / (sigma * sigma * 2))

centers = np.linspace(0, 10, 11)
sigmas = np.random.rand(11)/3 + 0.5 - 1/6
peaks = np.random.rand(11)/6 + 1
ys = np.vstack([gaussian(x0=centers[i], sigma=sigmas[i]) * peaks[i] for i in range(len(centers))])

fig = plt.figure(figsize=(4, 1))
colors = sns.color_palette("rainbow", len(centers))
ax = plt.axes()
for i, y in enumerate(ys):
    ax.plot(x, y, color=colors[i])
plt.savefig(r"C:\Users\12454\Desktop\Pre-Interviews\UNC interview figures\placefields.svg")
plt.show()
```
