```python
from mylib.statistic_test import *



def simplified_trace(trace: dict):
    trace_sim = {
        'MiceID': trace['MiceID'],
        'date': trace['date'],
        'maze_type': trace['maze_type'],
        'lap_begin_index': trace['lap_begin_index'],
        'lap_end_index': trace['lap_end_index'],
        'trace_folder': trace['p'],
        'processed_pos_new': trace['processed_pos_new'],
        'behav_time': trace['behav_time'],
        'behav_nodes': trace['behav_nodes'],
        'correct_pos': trace['correct_pos'],
        'correct_nodes': trace['correct_nodes'],
        'correct_time': trace['correct_time'],
        
        'Spikes_original': trace['Spikes'],
        'spike_nodes_original': trace['spike_nodes_original'],
        'ms_time': trace['ms_time'],
        'ms_speed_original': trace['ms_speed_original'],
        'RawTraces': trace['RawTraces'],
        'Ms': trace['Ms'],
        'n_neuron': trace['n_neuron'],
        'Spikes': trace['Spikes'],
        'ms_time_behav': trace['ms_time_behav'],
        'spike_nodes': trace['spike_nodes'],
        'rate_map_clear': trace['rate_map_clear'],
        'smooth_map_all': trace['smooth_map_all'],
        'smooth_map_fir': trace['smooth_map_fir'],
        'smooth_map_sec': trace['smooth_map_sec'],
        
        'old_map_clear': trace['old_map_clear'],
        'place_field_all': trace['place_field_all'],
        'SI_all': trace['SI_all'],
        'is_placecell': trace['is_placecell']
    }
    try:
        trace_sim['lap beg time'] = trace['lap beg time']
        trace_sim['lap end time'] = trace['lap end time']
    except:
        pass
    
    try:
        trace_sim['place_field_all_multiday'] = trace['place_field_all_multiday']
    except:
        trace_sim['place_field_all_multiday'] = trace['place_field_all']
    
    return trace_sim
    
idx = np.where(
    (np.isin(f1['Stage'], ['Stage 1', 'Stage 2'])) &
    (np.isin(f1['MiceID'], [10209, 10212, 10224, 10227, 10232, 10234]))
)[0]
for i in idx:
    if f1['include'][i] == 1:
        saved_path = join(
            r"D:\Data\Cross_maze",
            str(int(f1['MiceID'][i])),
            str(int(f1['date'][i])),
            f"session {int(f1['session'][i])}"
        )
        
        mkdir(saved_path)
        
        if not exists(f1['Trace File'][i]):
            print(f"Trace file does not exist: {f1['Trace File'][i]}")
            continue
        
        with open(join(
            r"E:\Data\Cross_maze",
            str(int(f1['MiceID'][i])),
            str(int(f1['date'][i])),
            f"session {int(f1['session'][i])}",
            "trace.pkl"
        ), 'rb') as handle:
            trace = pickle.load(handle)
    
        trace_sim = simplified_trace(trace)
        
        with open(join(saved_path, 'trace_sim.pkl'), 'wb') as handle:
            pickle.dump(trace_sim, handle)
```
